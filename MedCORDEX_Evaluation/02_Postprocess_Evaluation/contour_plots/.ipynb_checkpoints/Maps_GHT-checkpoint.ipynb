```python
import numpy as np
import os
import xarray as xr
import netCDF4
from netCDF4 import Dataset

yr = 1980
print("Data read started for the year", yr)
if 1980 <= yr <= 2014:
    directory = f'/data/inputs/METOCEAN/historical/model/atmos/ENEA/Med-CORDEX/historical/{yr}/'
else:
    directory = f'/data/inputs/METOCEAN/historical/model/atmos/ENEA/Med-CORDEX/projections/{yr}/'

file_psfc = Dataset(f"{directory}PSFC_{yr}.nc", mode="r")
PSFC = np.array(file_psfc.variables["PSFC"][:,:,:]) * 100 #Pa    #time, south_north, west_east

file_T2 = Dataset(f"{directory}T2_{yr}.nc", mode="r")
T2 = np.array(file_T2.variables["T2"][:,:,:]) #K    #time, south_north, west_east

file_RH2 = Dataset(f"{directory}RH2_{yr}.nc", mode="r")
RH2 = np.array(file_RH2.variables["RH2"][:,:,:]) #%    #time, south_north, west_east

file_T = Dataset(f"{directory}T_p_{yr}.nc", mode="r")
T = np.array(file_T.variables["T_p"][:,:,:,:]) #K: time, pressure, south_north, west_east
P = np.array(file_T.variables["pressure"][:]) * 100 #Pa

file_RH = Dataset(f"{directory}RH_p_{yr}.nc", mode="r")
RH = np.array(file_RH.variables["RH_p"][:,:,:,:]) #%: time, pressure, south_north, west_east

#file_geo = Dataset(f"/data/inputs/METOCEAN/historical/model/atmos/ENEA/Med-CORDEX/static/geo_em.d01.nc", mode="r")
#LSM = np.array(file_geo.variables["LANDMASK"][0,:,:]) #Time, south_north, west_east    "Landmask : 1=land, 0=water"
print("Data read completed")
```

```python
GHT = np.zeros(np.shape(T))

for t in range(0,1): #Time
    for lev in range(0,1): #Pr. lev
        ###### Finding values at different model levels #######
        # Method 1 for calculating qv
        #qv = RH[t,lev,:,:] / (26.3 * P[t,lev,:,:] / np.exp(17.67 * (T[t,lev,:,:]-273.16) / (T[t,lev,:,:]-29.65)))

        # Method 2 for calculating qv
        #rh = 1.E2 * (p*q/(q*(1.-0.622) + 0.622))/(611.2*np.exp(17.67*(t-273.15)/(T-29.65)))
        A = 611.2*np.exp(17.67*(T[t,lev,:,:]-273.15)/(T[t,lev,:,:]-29.65))
        RH_mod = np.where(RH[t,lev,:,:] == 0, 1, RH[t,lev,:,:]) # overwriting zero RH as 1 to avoid dividing by zero.
        qv2 = 0.622 / ((100 * P[lev] / (RH_mod * A)) - (1 - 0.622))
        rv2 = qv2/(1-qv2)
        Tv2 = T[t,lev,:,:] * (1 + (rv2/0.622)) / (1+rv2)
        #Tv2 = (1 + (0.61 * rv2)) * T[t,lev,:,:]

        ######### Finding  values at 2 m ##############
        A = 611.2*np.exp(17.67*(T2[t,:,:]-273.15)/(T2[t,:,:]-29.65))
        RH2_mod = np.where(RH2[t,:,:] == 0, 1, RH2[t,:,:]) # overwriting zero RH as 1 to avoid dividing by zero.
        qv1 = 0.622 / ((100 * PSFC[t,:,:] / (RH2_mod * A)) - (1 - 0.622))  #Ignoring the pressure difference existing between surface and 2m.

        rv1 = qv1/(1-qv1)
        Tv1 = (1 + (0.61 * rv1)) * T2[t,:,:]  #Virtual temperature at 2m.

        # Geometric height from hypsometric equation#
        Tv_mean = (Tv2 + Tv1) / 2
        Z1 = 2 # considering 2 meters for Z1
        #del = ((287 * Tv_mean /9.81) *np.log(PSFC[t,:,:]/P[lev]))
        Z = Z1 + ((287 * Tv_mean /9.81) *np.log(PSFC[t,:,:]/P[lev]))
        #### Calculating geopotential ####
        Re = 6378137 # Radius of earth in m
        GHT[t,lev,:,:] = Re * Z / (Re + Z)
```
