```python
import math, time, sys, os
import numpy as np
from numpy import sqrt,cos,ma
from mpl_toolkits.basemap import Basemap
import matplotlib as mpl
import matplotlib.pyplot as plt
import netCDF4 as nc
from   matplotlib.backends.backend_pdf import PdfPages
from   matplotlib.patches import Polygon
from   netCDF4 import Dataset as open_ncfile
from matplotlib.pyplot import figure
import scipy
from scipy.ndimage import convolve1d
from scipy.signal import savgol_coeffs, savgol_filter
from scipy.signal._savitzky_golay import _polyder
from scipy.interpolate import griddata
from matplotlib.ticker import (MultipleLocator, AutoMinorLocator)
from matplotlib import colors
import matplotlib.patches as patches
import datetime
from datetime import date, timedelta
import os
import subprocess
import cartopy.crs as ccrs
import cartopy.feature as cfeature
from sklearn.metrics import root_mean_squared_error, r2_score
from math import sqrt
import matplotlib.gridspec as gridspec
from mpl_toolkits.axes_grid1 import make_axes_locatable
import seaborn as sns
from sklearn.linear_model import LinearRegression
import pandas as pd
```

```python
############################################ Input #######################################
data_era5land="/work/cmcc/vr25423/Project/AdriaClimPlus/data/data_ERA5-Land/processed"
data_medcordex="/work/cmcc/vr25423/Project/AdriaClimPlus/data/data_MEDCORDEX/processed_SSP245_ERA5-Land"
##########################################################################################
```

```python
######################### Temperature ############################
yr_v = ['1985','1986','1987','1988','1989','1990', '1991', '1992', '1993', '1994', '1995', '1996', '1997', '1998', '1999',
        '2000', '2001', '2002', '2003', '2004', '2005', '2006', '2007', '2008', '2009',
        '2010', '2011', '2012', '2013', '2014']

t2_era5land_winter_m = []
t2_era5land_spring_m = []
t2_era5land_summer_m = []
t2_era5land_autumn_m = []
t2_medcordex_remap_winter_m = []
t2_medcordex_remap_spring_m = []
t2_medcordex_remap_summer_m = []
t2_medcordex_remap_autumn_m = []

for year in yr_v:
    ### winter ###
    t2_era5land = open_ncfile(str(data_era5land)+"/ERA5-Land_"+str(year)+"_t2m_winter.nc","r")
    t2_medcordex = open_ncfile(str(data_medcordex)+"/MEDCORDEX_"+str(year)+"_t2m_remap_winter.nc","r")
    t2_era5land_winter = t2_era5land.variables["t2m"][0,:,:]-273.15
    t2_era5land_winter_m.append(t2_era5land_winter)
    t2_medcordex_winter=t2_medcordex.variables["T2"][0,:,:]-273.15
    t2_medcordex_remap_winter_m.append(np.where(np.isnan(t2_era5land_winter), np.nan, t2_medcordex_winter))
    
    
    ### spring ###
    t2_era5land = open_ncfile(str(data_era5land)+"/ERA5-Land_"+str(year)+"_t2m_spring.nc","r")
    t2_medcordex = open_ncfile(str(data_medcordex)+"/MEDCORDEX_"+str(year)+"_t2m_remap_spring.nc","r")
    t2_era5land_spring = t2_era5land.variables["t2m"][0,:,:]-273.15
    t2_era5land_spring_m.append(t2_era5land_spring)
    t2_medcordex_spring=t2_medcordex.variables["T2"][0,:,:]-273.15
    t2_medcordex_remap_spring_m.append(np.where(np.isnan(t2_era5land_spring), np.nan,t2_medcordex_spring))
    
    ### summer ###
    t2_era5land = open_ncfile(str(data_era5land)+"/ERA5-Land_"+str(year)+"_t2m_summer.nc","r")
    t2_medcordex = open_ncfile(str(data_medcordex)+"/MEDCORDEX_"+str(year)+"_t2m_remap_summer.nc","r")
    t2_era5land_summer = t2_era5land.variables["t2m"][0,:,:]-273.15
    t2_era5land_summer_m.append(t2_era5land_summer)
    t2_medcordex_summer=t2_medcordex.variables["T2"][0,:,:]-273.15
    t2_medcordex_remap_summer_m.append(np.where(np.isnan(t2_era5land_summer), np.nan,t2_medcordex_summer))
    
    
    ### autumn ###
    t2_era5land = open_ncfile(str(data_era5land)+"/ERA5-Land_"+str(year)+"_t2m_autumn.nc","r")
    t2_medcordex = open_ncfile(str(data_medcordex)+"/MEDCORDEX_"+str(year)+"_t2m_remap_autumn.nc","r")
    t2_era5land_autumn = t2_era5land.variables["t2m"][0,:,:]-273.15
    t2_era5land_autumn_m.append(t2_era5land_autumn)
    t2_medcordex_autumn=t2_medcordex.variables["T2"][0,:,:]-273.15
    t2_medcordex_remap_autumn_m.append(np.where(np.isnan(t2_era5land_autumn), np.nan,t2_medcordex_autumn))

t2_era5land_winter_m = np.array(t2_era5land_winter_m).flatten()
t2_era5land_spring_m = np.array(t2_era5land_spring_m).flatten()
t2_era5land_summer_m = np.array(t2_era5land_summer_m).flatten()
t2_era5land_autumn_m = np.array(t2_era5land_autumn_m).flatten()

t2_medcordex_remap_winter_m = np.array(t2_medcordex_remap_winter_m).flatten()
t2_medcordex_remap_spring_m = np.array(t2_medcordex_remap_spring_m).flatten()
t2_medcordex_remap_summer_m = np.array(t2_medcordex_remap_summer_m).flatten()
t2_medcordex_remap_autumn_m = np.array(t2_medcordex_remap_autumn_m).flatten()
```

```python

# Define seasons and corresponding temperature data
seasons = ['Winter', 'Spring', 'Summer', 'Autumn']
era5land_temps = [t2_era5land_winter_m, t2_era5land_spring_m, t2_era5land_summer_m, t2_era5land_autumn_m]
medcordex_temps = [t2_medcordex_remap_winter_m, t2_medcordex_remap_spring_m, 
                   t2_medcordex_remap_summer_m, t2_medcordex_remap_autumn_m]

fig, axes = plt.subplots(2, 2, figsize=(10, 10))  # Create 2x2 subplots
axes = axes.flatten()  # Flatten to loop over easily

for i, season in enumerate(seasons):
    # Create DataFrame
    df = pd.DataFrame({'ERA5-Land Temperature (C)': era5land_temps[i].flatten(), 
                       'Med-CORDEX Temperature (C)': medcordex_temps[i].flatten()})
    
    # Remove NaNs
    df = df.dropna()

    # Compute xlim and ylim dynamically
    x_min, x_max = df['ERA5-Land Temperature (C)'].min(), df['ERA5-Land Temperature (C)'].max()
    y_min, y_max = df['Med-CORDEX Temperature (C)'].min(), df['Med-CORDEX Temperature (C)'].max()

    def round_to_nearest_5(value):
        return round(value / 5) * 5
    
    x_min = round_to_nearest_5(x_min)
    x_max = round_to_nearest_5(x_max)
    y_min = round_to_nearest_5(y_min)
    y_max = round_to_nearest_5(y_max)
    

    # Ensure symmetry in axes for fair comparison
    min_limit = min(x_min, y_min) - 5
    max_limit = max(x_max, y_max) + 5

    # Scatter plot
    axes[i].scatter(df['ERA5-Land Temperature (C)'], df['Med-CORDEX Temperature (C)'], 
                    color='blue', alpha=0.1)

    # Fit regression model
    X = df['ERA5-Land Temperature (C)'].values.reshape(-1, 1)
    y = df['Med-CORDEX Temperature (C)'].values

    reg_model = LinearRegression()
    reg_model.fit(X, y)

    # Generate extended X values
    X_extended = np.linspace(X.min(), X.max(), 100).reshape(-1, 1)
    y_extended = reg_model.predict(X_extended)

    # Plot regression line
    axes[i].plot(X_extended, y_extended, color='red', linewidth=2, linestyle='--')

    # Reference 1:1 line (dynamically adjusted)
    axes[i].plot([min_limit, max_limit], [min_limit, max_limit], color='black', linewidth=1.5)

    # Labels and titles
    axes[i].set_xlabel('ERA5-Land Temperature (C)', fontsize=12)
    axes[i].set_ylabel('Med-CORDEX Temperature (C)', fontsize=12)
    axes[i].set_title(season, fontsize=14)

    # Adjust ticks
    axes[i].tick_params(axis='both', labelsize=12)

    # Set limits dynamically
    axes[i].set_xlim(min_limit, max_limit)
    axes[i].set_ylim(min_limit, max_limit)

    # Grid
    axes[i].grid(True, linestyle='-', alpha=0.3)

    axes[i].text(min_limit + 2, max_limit - 2, 'Land', fontsize=16, color='black', ha='left', va='top', alpha=0.7)

# Adjust layout
plt.tight_layout()
plt.show()
```
